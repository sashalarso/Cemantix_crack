```python
import requests

with open("liste.de.mots.francais.frgut.txt","r") as fichier:
 
    lignes = fichier.readlines()

   
    liste_lignes = []

    
    for ligne in lignes:
        liste_lignes.append(ligne.strip()) 
i=0
max_score=0
for word in liste_lignes:
    resp =requests.post("https://cemantix.certitudes.org/score",data={"word":word}).json()
    if "score" in resp:
        if resp["score"]>max_score:
            max_score=resp["score"]
            print("max partiel : " + str(resp["score"]) + " mot : " + word)
        if resp["score"]>0.995:
            print("Le mot est trouvé c'est : " + word)
            break
    if i%1000==0:
        print(i)
    i+=1

```
